```python
from pyspark.sql.functions import col, trim, upper, to_date
```

```python
customers_silver_df = customers_df \
    .withColumn("signup_date", to_date(col("signup_date"))) \
    .withColumn("country", upper(trim(col("country"))))
```

```python
customers_silver_df.printSchema()
customers_silver_df.show(5)
```

```python
from pyspark.sql.functions import when
```

```python
products_silver_df = products_df \
    .withColumn("category", upper(trim(col("category")))) \
    .withColumn(
        "price",
        when(col("price").isNull(), 0).otherwise(col("price"))
    )
```

```python
orders_silver_df = orders_df \
    .withColumn("order_date", to_date(col("order_date"))) \
    .withColumn("status", upper(trim(col("status"))))
```

```python
order_items_silver_df = order_items_df \
    .filter(col("quantity") > 0) \
    .withColumn(
        "unit_price",
        when(col("unit_price").isNull(), 0).otherwise(col("unit_price"))
    )
```

```python
from pyspark.sql.functions import expr, upper, trim, col

orders_silver_df = orders_df \
    .withColumn(
        "order_date",
        expr("try_cast(order_date as date)")
    ) \
    .withColumn(
        "status",
        upper(trim(col("status")))
    )
```

```python
orders_silver_df \
    .where(col("order_date").isNull()) \
    .show()
```

```python
orders_silver_df \
    .where(col("order_date").isNull()) \
    .count()
```

```python
orders_invalid_dates_df = orders_silver_df \
    .filter(col("order_date").isNull())
```

```python
orders_silver_clean_df = orders_silver_df \
    .filter(col("order_date").isNotNull())
```

```python
orders_silver_clean_df \
    .select("order_date") \
    .where(col("order_date").isNull()) \
    .count()
```

```python
dim_customer_df = customers_silver_df.select(
    "customer_id",
    "first_name",
    "last_name",
    "email",
    "country",
    "signup_date"
).dropDuplicates(["customer_id"])
```

```python
dim_product_df = products_silver_df.select(
    "product_id",
    "product_name",
    "category",
    "price"
).dropDuplicates(["product_id"])
```

```python
dim_store_df = stores_df.select(
    "store_id",
    "store_name",
    "city"
).dropDuplicates(["store_id"])
```

```python
from pyspark.sql.functions import year, month, dayofmonth
```

```python
dim_date_df = orders_silver_clean_df \
    .select("order_date") \
    .dropDuplicates() \
    .withColumn("year", year(col("order_date"))) \
    .withColumn("month", month(col("order_date"))) \
    .withColumn("day", dayofmonth(col("order_date")))
```

```python
fact_sales_df = orders_silver_clean_df \
    .join(order_items_silver_df, "order_id") \
    .join(products_silver_df, "product_id") \
    .select(
        col("order_id"),
        col("order_date"),
        col("customer_id"),
        col("store_id"),
        col("product_id"),
        col("quantity"),
        col("unit_price"),
        (col("quantity") * col("unit_price")).alias("total_amount")
    )
```

# Ventas por mes

---



```python
spark.sql("""
SELECT
    d.year,
    d.month,
    SUM(f.total_amount) AS total_sales
FROM fact_sales f
JOIN dim_date d
    ON f.order_date = d.order_date
GROUP BY d.year, d.month
ORDER BY d.year, d.month
""").show()
```

# Productos con mas ingresos

```python
spark.sql("""
SELECT
  p.product_name,
  SUM(f.total_amount) AS revenues
FROM fact_sales f
JOIN dim_product p
  ON f.product_id = p.product_id
GROUP BY p.product_name
ORDER BY revenues DESC
LIMIT 10
""").show()
```

# Ventas por pas

```python
spark.sql("""
SELECT
    c.country,
    SUM(f.total_amount) AS total_sales
FROM fact_sales f
JOIN dim_customer c
    ON f.customer_id = c.customer_id
GROUP BY c.country
ORDER BY total_sales DESC
""").show()
```

#  Venta por tienda y mes

```python
spark.sql("""
SELECT
    s.store_name,
    d.year,
    d.month,
    SUM(f.total_amount) AS total_sales
FROM fact_sales f
JOIN dim_store s
    ON f.store_id = s.store_id
JOIN dim_date d
    ON f.order_date = d.order_date
GROUP BY s.store_name, d.year, d.month
ORDER BY s.store_name, d.year, d.month
""").show()
```
